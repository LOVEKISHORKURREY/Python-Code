# how to create dataframe

```python
d = pd.DataFrame([[2, 3], [5, 6], [8, 9]],
     index=['cobra', 'viper', 'sidewinder'],
     columns=['max_speed', 'shield'])
```

```python
d
```

```python
d.loc['viper']
```

```python
d.loc[['viper', 'sidewinder']]
```

```python
d
```

```python
d.loc[d['shield'] > 6]
#Conditional that returns a boolean Series
```

```python
d.loc[d['shield'] > 6, ['max_speed']]
#Conditional that returns a boolean Series with column labels specified
```

```python
# Set value for all items matching the list of labels
d.loc[['viper', 'sidewinder'], ['shield']] = 30
```

```python
d
```

```python
# Set value for an entire row
d.loc['cobra'] =10
d
```

```python
# Set value for an entire column:

d.loc[:, 'max_speed'] = 30
d
```

```python
d = pd.DataFrame([[2, 3], [5, 6], [8, 9]],
     index=[2, 3, 4], columns=['max_speed', 'shield'])
d
#Getting values on a DataFrame with an index that has integer labels
```

```python
#Slice with integer labels for rows. As mentioned above, note that both the start and stop of the slice are included:

d.loc[2:4]
```

```python
# Getting values with a MultiIndex
# A number of examples using a DataFrame with a MultiIndex:

tuples = [
   ('cobra', 'm1'), ('cobra', 'm2'),
   ('sidewinder', 'm1'), ('sidewinder', 'm2'),
   ('viper', 'm2'), ('viper', 'm3')
]
```

```python
index = pd.MultiIndex.from_tuples(tuples)
values = [[14, 4], [0, 6], [20, 30],
         [2, 4], [5, 1], [26, 46]]
```

```python
d = pd.DataFrame(values, columns=['max_speed', 'shield'], index=index)
d
```

```python
d.loc['cobra']
```

```python
d.loc[('cobra', 'm2')]
```

```python
d.loc['cobra', 'm1']
```

```python
d.loc[[('cobra', 'm2')]]
```

```python
#Slice from index tuple to index tuple:

d.loc[('cobra', 'm1'):('viper', 'm2')]
```

```python
db = pd.DataFrame([[0, 3, 4], [0, 5, 2], [20, 30, 40]],
                  index=[1, 2, 3], columns=['P', 'Q', 'R'])
db
```

```python
#DataFrame - at property

#The at property is used to access a single value for a row/column label pair.

#Similar to loc, in that both provide label-based lookups. 
#Use at if you only need to get or set a single value in a DataFrame or Series.

#Syntax: DataFrame.at
```

```python
#Get value at specified row/column pair:

db.at[1, 'Q']
```

```python
#Set value at specified row/column pair:

db.at[1, 'Q'] = 20
db.at[1, 'Q']
```

```python
#Get value within a Series:

db.loc[2].at['Q']
```

```python
mydict = [{'p': 2, 'q': 3, 'r': 4, 's': 5},
          {'p': 200, 'q': 300, 'r': 400, 's': 500},
          {'p': 2000, 'q': 3000, 'r': 4000, 's': 5000 }]

dm = pd.DataFrame(mydict)
dm
```

```python
#With a list of integers.

dm.iloc[[0]]
```

```python
type(dm.iloc[[0]])
```

```python
dm.iloc[[0, 2]]
```

```python
#With a slice object:

dm.iloc[:3]
```

```python
#With slice objects:

dm.iloc[1:3, 0:3]
```
